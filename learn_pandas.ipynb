```python
# rate efficiency class
cars["efficiency_class"] = cars.efficiency_wh_per_km.map(lambda s: "High Efficieny" if s < 160 else ("Moderate Efficiency" if s < 200 else "Low Efficiency"))
cars.loc[:,["brand", "model", "efficiency_wh_per_km", "efficiency_class" ]]



# rate performance score

def perf_score(row):
  drivetrain_factor = 0
  match (row.drivetrain):
    case "AWD":
      drivetrain_factor = 1.1
    case "RWD":
      drivetrain_factor = 1.0
    case "FWD":
      drivetrain_factor = 0.8
    case _:
      drivetrain_factor = 1.0

  return (row.torque_nm / row.acceleration_0_100_s) * drivetrain_factor



cars["performance_score"] = cars.apply(perf_score, axis=1)


cars.loc[:, ["brand", "model", "performance_score"]].sort_values(by="performance_score")
```
